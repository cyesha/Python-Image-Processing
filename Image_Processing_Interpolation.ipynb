# Interpolation to np.eye(3)

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = np.eye(3)
img[1,2] = 1
```

```python
plt.imshow(img)
plt.title("No Interpolation")
plt.show()
```

```python
no_interpolation = cv2.resize(img,(1000,1000), interpolation = None)
plt.imshow(no_interpolation)
plt.title("No Interpolation (Resized)")
plt.show()
```

```python
nearest_interpolation = cv2.resize(img,(1000,1000), interpolation = cv2.INTER_NEAREST)
plt.imshow(nearest_interpolation)
plt.title("Nearest Neighbor")
plt.show()
```

```python
linear_interpolation = cv2.resize(img,(1000,1000), interpolation = cv2.INTER_LINEAR)
plt.imshow(linear_interpolation )
plt.title("Bilinear Interpolation")
plt.show()
```

```python
bicubic_interpolation = cv2.resize(img,(1000,1000), interpolation = cv2.INTER_CUBIC)
plt.imshow(bicubic_interpolation )
plt.title("Bicubic Interpolation")
plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(15, 5))
axes[0].imshow(nearest_interpolation)
axes[0].set_title("Nearest Neighbor Interpolation")
axes[1].imshow(linear_interpolation)
axes[1].set_title("Bilinear Interpolation")
axes[2].imshow(bicubic_interpolation)
axes[2].set_title("Bicubic Interpolation")
axes[3].imshow(img)
axes[3].set_title("Original Image")
axes[4].imshow(no_interpolation)
axes[4].set_title("Resized Image")
```

Explain the three common interpolation methods


* Nearest Neighbor
* Bilinear
* Bicubic

Discuss the output after applying interpolation in resizing the image

Answer:

The three common interpolation methods are:

Nearest Neighbor Interpolation:
The simplest interpolation method, nearest neighbor interpolation selects the value of the nearest pixel to the interpolated point to estimate the new pixel value. The resulting image can have a blocky appearance due to the pixelated nature of the method, especially when upscaling the image.

Bilinear Interpolation:
Bilinear interpolation uses the values of the four nearest pixels to the interpolated point to estimate the new pixel value. It takes a weighted average of the four pixels based on the distance between the interpolated point and each pixel. This method results in smoother images with less visible pixelation.

Bicubic Interpolation:
Bicubic interpolation uses the values of the sixteen nearest pixels to the interpolated point to estimate the new pixel value. It takes a weighted average of the sixteen pixels based on their distance from the interpolated point, and also uses the gradient information to smooth the image. This method produces the smoothest images of the three methods but can be computationally expensive.

The output after applying interpolation in resizing the image:
Nearest Neighbor interpolation produces a blocky image with visible pixelation due to its simplistic approach. Bilinear interpolation produces a smoother image that is generally acceptable for most purposes. Bicubic interpolation produces the smoothest image of the three methods but can be computationally expensive. In the example above, the one that gives closest to the original is Nearest Neighbor Interpolation.

```python
def my_show(ax, img, title=None, interpolation='bicubic', **kwargs):
    ' helper to display an image on an axes without grid/spine '
    ax.imshow(img, interpolation = interpolation, **kwargs)
    ax.axis('off')
    if title:
        ax.set_title(title)

def my_gshow(ax, img, title=None, cmap='gray', interpolation='bicubic', **kwargs):
    ' helper to display an image, in grayscale, on an axes without grid/spine '
    my_show(ax, img, title=title, cmap='gray', interpolation=interpolation, **kwargs)
```
